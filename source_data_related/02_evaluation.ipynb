```python
import os

# Define the directory containing the log files
log_dir = 'log/log/0_other_model'

# Initialize a dictionary to store the data
learning_data = {}

# Iterate over all files in the directory
for filename in os.listdir(log_dir):
    if filename.endswith('.log'):
        model_name = filename.split('.')[0]
        with open(os.path.join(log_dir, filename), 'r') as file:
            lines = file.readlines()
            accuracies = []
            for line in lines:
                if 'val ap:' in line:
                    val_ap = float(line.split('val ap:')[1].split(',')[0].strip())
                    accuracies.append(val_ap)
            learning_data[model_name] = accuracies

learning_data["DyRep"]=
# Plot the learning curves
plt.figure(figsize=(10, 6))
for model_name, data in learning_data.items():
    plt.plot(data, label=model_name)

plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Learning Curves')
plt.legend()
plt.show()
```
